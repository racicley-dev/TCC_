```python
import pandas as pd
import os
```

```python
#Gerar um arquivo unificado
pasta='arquivos_csv_unidos/'
arquivos = [pd.read_csv(pasta+arquivo) for arquivo in os.listdir(pasta) if os.path.isfile(os.path.join(pasta, arquivo))]
#arquivos
```

```python
#Load and Clean
df = pd.concat(arquivos, ignore_index=True)
df = df[['PMID','Title', 'abstract']]
df = df.drop_duplicates()
```

```python
df.rename(columns={'Title':'TITLE', 'abstract':'ABSTRACT'}, inplace=True)
```

```python
#Removendo os abstracts nulos
df = df[df['ABSTRACT'].isnull() == False]
df
```

```python
#Abstract iguais
df[df['ABSTRACT'].duplicated() == True].sort_values('ABSTRACT', ascending=True)
```

```python
#Abstract iguais
df = df[df['TITLE'].duplicated() == False]
```

```python
#Salve em csv
df.to_csv('csv_buscas.csv')
```
