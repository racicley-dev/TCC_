```python
#Load and Clean
#df = pd.concat(arquivos, ignore_index=True)
df = pd.read_csv('csv_buscas_encoded.csv')
df = df[['PMID','TITLE', 'ABSTRACT']]
df = df.drop_duplicates()
```

```python
#Removendo os abstracts nulos
df = df[df['ABSTRACT'].isnull() == False]
df
```

```python
#Abstract iguais
df[df['ABSTRACT'].duplicated() == True].sort_values('ABSTRACT', ascending=True)
```

```python
#Abstract iguais
df = df[df['TITLE'].duplicated() == False]
```

```python
df
```

```python
df["TITLE"] = df["TITLE"].apply(lambda x: x.decode('utf-8'))
```

```python
df["ABSTRACT"] = df["ABSTRACT"].apply(lambda x: x.decode('utf-8'))
```

```python
df = df[df['PMID'] != 25381313]
df
```
